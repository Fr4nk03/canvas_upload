```python
# TOTAL SCORE including EC
for index, row in hw_score.iterrows():
  netid = row["Email"].split("@")[0]
  hw_score.loc[index, "SID"] = netid
  score_before_ec = row["1: Q1 (10.0 pts)"] + row["2: Q2 (10.0 pts)"] + row["3: Q3 (10.0 pts)"] + row["4: Q4 (10.0 pts)"] + row["5: Q5 (10.0 pts)"]
  hw_score.loc[index, "Total Score Out of 50"] = score_before_ec

  # ec = row["6: Q6 (3.0 pts)"]
  ec = row["6: Q6 (2.0 pts)"]
  # hw_score.loc[index, "Total Score without EC"] = score
  hw_score.loc[index, "EC"] = ec
hw_score["SIS User ID"] = hw_score["SID"]
hw_score.head()
```
