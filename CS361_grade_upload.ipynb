```python
import pandas as pd
import numpy as np
```

```python
hw9_score = pd.read_csv("HW9_scores.csv")
print(hw9_score.shape)
hw9_score.head()
```

```python
for index, row in hw9_score.iterrows():
  netid = row["Email"].split("@")[0]
  hw9_score.loc[index, "SID"] = netid
  score_before_ec = row["1: Q1 (10.0 pts)"] + row["2: Q2 (10.0 pts)"] + row["3: Q3 (10.0 pts)"] + row["4: Q4 (10.0 pts)"] + row["5: Q5 (10.0 pts)"]
  hw9_score.loc[index, "Total Score Out of 50"] = score_before_ec
hw9_score["SIS User ID"] = hw9_score["SID"]
hw9_score.head()
```

```python
canvas_grade = pd.read_csv("canvas_gradebook.csv")
print(canvas_grade.shape)
canvas_grade.head()
# print(canvas_grade.loc[244, "HW9 (1050916)"])
```

```python
# print(hw9_score["Total Score"][0])
for index, row in canvas_grade.iterrows():
  if (index != 0 and index != canvas_grade.shape[0] - 1):
    netid = canvas_grade["SIS User ID"][index]
    # print("Netid: " + str(netid))
    score = hw9_score[hw9_score["SIS User ID"] == netid]["Total Score Out of 50"]
    if not score.empty:
      # print(score)
      # print("Score: " + str(score.iloc[0]))
      canvas_grade.loc[index, "HW9 (1050916)"] = score.iloc[0]

canvas_grade.head()
# print(canvas_grade.loc[244, "HW9 (1050916)"])
```

```python
canvas_grade["HW9 (1050916)"]
canvas_grade.loc[:canvas_grade.shape[0] - 2, "HW9 (1050916)"] = canvas_grade.loc[:canvas_grade.shape[0] - 2, "HW9 (1050916)"].fillna(0.0)

canvas_grade["HW9 (1050916)"]
```
