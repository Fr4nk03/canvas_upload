```python
import pandas as pd
import numpy as np
```

### Define relevant folder paths, date-time for quiz, and quiz points

### Due Dates

```python
due_dates = {"Quiz 3": "2024-09-22T23:59",
"Quiz 4": "2024-10-05T23:59",
"Quiz 5": "2024-10-19T23:59"}

# Name of the quiz columns in the grade book
quiz_columns = {"Quiz 3":"Quiz3  (1050865)",\
                "Quiz 4": "Quiz4  (1050872)",\
                "Quiz 5": "Quiz5  (1050870)"}

quiz_str = "Quiz 5"
quiz_attempt_one_fp = "data/CS_361_Fa2024_Q5_instances.csv"
quiz_attempt_two_fp = "data/CS_361_Fa2024_Q5_instances_2.csv"
```

```python
# Year, Month, Day, Hour, Minute
due_date_str = due_dates[quiz_str]

# canvas quiz string
quiz_str = quiz_columns[quiz_str]

#Output fp
output_fp = "data/CS_361_Fa2024_Q4_grades.csv"

# Canvas CSV
canvas_fp = "data/2024-12-08T1527_Grades-cs_361_120248_237374.csv"
```

# Data Preparation

```python
quiz_percentage1 = pd.read_csv(quiz_attempt_one_fp)
max_point = quiz_percentage1.loc[1, "Max points"]
print(f"Max Point: {max_point}")
quiz_percentage1 = quiz_percentage1.rename(columns={'Score (%)': 'First Score (%)'})
quiz_percentage1 = quiz_percentage1.rename(columns={'Username': 'SID'})

quiz_percentage1 = quiz_percentage1[["SID", "Started", "First Score (%)"]]
print(quiz_percentage1.shape)
quiz_percentage1.head()
```

```python
for index, row in quiz_percentage1.iterrows():
  # netid = row["UID"].split("@")[0]
  # quiz_percentage1.loc[index, "SID"] = netid
  quiz_percentage1.loc[index, "First Score (%)"] = min(100, row["First Score (%)"])  # cap at 100

quiz_percentage1.head()
```

```python
quiz_percentage2 = pd.read_csv(quiz_attempt_two_fp)
quiz_percentage2 = quiz_percentage2.rename(columns={'Score (%)': 'Second Score (%)'})
quiz_percentage2 = quiz_percentage2.rename(columns={'Username': 'SID'})
quiz_percentage2 = quiz_percentage2[["SID", "Second Score (%)"]]
print(quiz_percentage2.shape)
quiz_percentage2.head()
```

```python
for index, row in quiz_percentage2.iterrows():
  # netid = row["UID"].split("@")[0]
  # quiz_percentage2.loc[index, "SID"] = netid
  quiz_percentage2.loc[index, "Second Score (%)"] = min(100, row["Second Score (%)"])  # cap at 100

quiz_percentage2.head()
```

**Merge Two Tables**

```python
for index, row in quiz_percentage1.iterrows():
  SID = row["SID"]
  if SID not in quiz_percentage2['SID'].values:
    quiz_percentage2.loc[len(quiz_percentage2)] = [SID, np.nan]

print(quiz_percentage2.shape)
quiz_percentage2
```

```python
quiz_average = pd.merge(quiz_percentage1, quiz_percentage2, on="SID")
quiz_average
```

# Calculations

**Average two attempts properly**


*   *Take first score as Average if no second attemp*



```python
quiz_average['Average'] = quiz_average.apply(
    lambda row: row['First Score (%)'] if pd.isna(row['Second Score (%)'])
    else (row['First Score (%)'] + row['Second Score (%)']) / 2,
    axis=1
)
quiz_average
```

**Bonus**

```python
import pytz
from datetime import datetime

# Define the due date with timezone
# due_date_str is defined at the top
due_date = pytz.timezone('America/Chicago').localize(datetime.strptime(due_date_str, '%Y-%m-%dT%H:%M'))
print(f"Due date: {due_date}")

# Comparison
for index, row in quiz_average.iterrows():
  start_date = pd.to_datetime(row["Started"])
  # print(f"Start Date: {start_date}")
  is_on_time = start_date < due_date
  quiz_average.loc[index, "Is_On_Time"] = is_on_time
  updated_percentage = row["Average"] * (1.05) if is_on_time == True else row["Average"]
  quiz_average.loc[index, "Bonus Added"] = updated_percentage
  print(row) # For some reason the row is not updated unless i do this?
  quiz_average.loc[index, "Raw Points"] = updated_percentage * max_point / 100
  print(row) # For some reason the row is not updated unless i do this?


quiz_average
```

```python
quiz_average.to_csv(output_fp)
```
